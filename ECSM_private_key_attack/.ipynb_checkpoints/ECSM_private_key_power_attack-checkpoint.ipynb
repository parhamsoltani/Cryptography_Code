## Generating a random private key and sample power traces based on its hamming wights

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

num_samples = 1000              # Number of power trace samples
key_bits = 256                  # Private key length (256 bits)
noise_variance = 0.05           # Gaussian noise variance

# Generate random private key (binary) without fixed seed
private_key = np.random.randint(0, 2, key_bits)  # Simulated private key

# Simulate power consumption based on Hamming weight
power_trace = np.zeros(num_samples)
for i in range(key_bits):
    hamming_weight = np.sum(private_key[:i+1])  # Calculate cumulative Hamming weight
    power_trace[i * (num_samples // key_bits):(i+1) * (num_samples // key_bits)] = hamming_weight

# Add Gaussian noise to simulate real conditions
power_trace += np.random.normal(0, np.sqrt(noise_variance), num_samples)

# Display simulated power trace
plt.figure(figsize=(12, 6))
plt.plot(power_trace)
plt.title("Simulated Power Trace for Scalar Multiplication (ECC)")
plt.xlabel("Sample Index")
plt.ylabel("Power Consumption")
plt.grid(True)
plt.show()

# Save power trace and private key to file
trace_data = pd.DataFrame({
    'Sample_Index': np.arange(num_samples),
    'Power_Consumption': power_trace
})

# File paths for saving
trace_path = 'simulated_power_trace.csv'
key_path = 'simulated_private_key.txt'

# Save power trace and private key
trace_data.to_csv(trace_path, index=False)
with open(key_path, 'w') as f:
    f.write(''.join(map(str, private_key)))

print(f"Power Trace saved as: {trace_path}")
print(f"Private Key saved as: {key_path}")
```

## Attacking the Elliptic Curve scalar multiplication to extract the private key

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import pywt
from scipy.signal import savgol_filter, butter, filtfilt
from scipy import signal


# Load the data
trace_data = pd.read_csv('simulated_power_trace.csv')
samples = trace_data.values.flatten()

with open('simulated_private_key.txt', 'r') as f:
    real_private_key = f.read().strip()
    

def conditional_variables_extraction(traces):
    """
    Implementation of Algorithm 5: Conditional Variables Extraction
    
    Parameters:
    - traces: Array T[i,j] where 1  i  n, 1  j  m
    
    Returns:
    - List containing conditions
    """
    V = np.var(traces, axis=0)  # V[j] along the x-axis
    E = np.mean(traces, axis=0)  # E[j] along the x-axis
    count = 0
    pol = []
    threshold = 0
    
    # Finding positions where variance exceeds mean
    for j in range(traces.shape[1]):
        if V[j] > E[j]:
            count += 1
            pol.append(j)
            threshold += E[j]
    
    # Computing conditions
    conditions = []
    for i in range(traces.shape[0]):
        sum_val = 0
        for j in pol:
            sum_val += traces[i][j]
        conditions.append(sum_val > threshold)
    
    return conditions

def table_indices_extraction(traces, G):
    """
    Implementation of Algorithm 6: Table Indices Extraction
    
    Parameters:
    - traces: Array T[i,j] where 0  i < n, 1  j  m
    - G: Base point
    
    Returns:
    - List containing k indices
    """
    n = traces.shape[0]
    indices = []
    
    for i in range(n):
        j = []
        max_val = float('-inf')
        
        for k in range(8):
            # Compute ((kx - x0)x0)  (2k + 1)16G
            val = ((k * 16) - 16) * 16
            # Add correlation computation here based on Eq. (2)
            if val > max_val:
                max_val = val
                j = k
        
        indices.append(j)
    
    return indices


def private_key_reconstruction(clist, jlist, p256):
    """
    Implementation of Algorithm 7: Private Key Reconstruction
    
    Parameters:
    - clist: List from Algorithm 5
    - jlist: List from Algorithm 6
    - p256: Prime modulus
    
    Returns:
    - Reconstructed private key
    """
    a = 2 * jlist[0] + 1
    
    for i in range(1, 64):
        if clist[i]:
            a = -a
        if jlist[i]:
            a = a + (2 * jlist[i] + 1) * 16
        
    return a % p256

private_key_var = real_private_key
    
def elliptic_curve_scalar_mult(k, T):
    """
    Implementation of Algorithm 1: Elliptic Curve Scalar Multiplication
    
    Parameters:
    - k: Scalar
    - T: Precomputed table T[i][j]
    
    Returns:
    - Result of scalar multiplication
    """
    
    a = k + 2**256
    
    if a % 2 == 0:
        a = a - p256
        
    if a == 0:
        return float('inf')
    
    b = a & 0x1F
    b = b ^ ((b >> 4) - 1) & 0xF
    R = None  # J2A(T[0][b >> 1])
    
    for i in range(1, 63):
        a = a >> 4
        b = a & 0x1F
        b = b ^ ((b >> 4) - 1) & 0xF
        
        if -b & 1:
            R = -R
        R = T[i][b >> 1] + R
        if -(a >> 4) & 1:
            R = -R
            
    return R  # A2J(R)


def field_multiplication(a, b, p256):
    """
    Implementation of Algorithm 4: Field Multiplication Modulo p256
    
    Parameters:
    - a, b: 256-bit integers
    - p256: Prime modulus
    
    Returns:
    - Result of multiplication modulo p256
    """
    t = 0
    c = [0] * 18
    
    # First loop
    for i in range(9):
        for j in range(i + 1):
            t = t + a[j] * b[i-j]
        c[i] = t & 0x1FFFFFFF
        t = t >> 29
    
    # Second loop
    for i in range(9, 17):
        for j in range(i-8, 9):
            t = t + a[j] * b[i-j]
        c[i] = t & 0x1FFFFFFF
        t = t >> 29
    
    c[17] = t
    
    return c  # After fast modular reduction


# Parameters
num_samples = len(samples)
key_bits = len(real_private_key)
initial_guess = private_key_var[:10]
window_size = num_samples // key_bits


# Step 1: Advanced Noise Reduction ---

# Apply Savitzky-Golay filter for smoothing
filtered_samples = savgol_filter(samples, window_length=11, polyorder=2)

# Wavelet Denoising
coeffs = pywt.wavedec(filtered_samples, 'haar', level=3)  
coeffs[-1] = np.zeros_like(coeffs[-1])  # Remove high-frequency noise
filtered_samples = pywt.waverec(coeffs, 'haar')

# Assume a sampling rate (fs) if not known
fs = 1000  # Example: 1000 samples per second (adjust based on your data)
nyquist = 0.5 * fs

# Normalize cutoff frequencies
low = 0.1 / nyquist
high = 10 / nyquist

# Design the Butterworth bandpass filter
b, a = butter(3, [low, high], btype='band')
filtered_samples = filtfilt(b, a, filtered_samples)

# Step 2: Improved CPA with Guided Correction 

guessed_key = initial_guess


for bit_index in range(10, key_bits):  # Start from the 11th bit onward
    start_idx = bit_index * window_size
    end_idx = (bit_index + 1) * window_size
    window = filtered_samples[start_idx:end_idx]

    # Extract features
    mean_power = np.mean(window)
    variance_power = np.var(window)
    signature = np.sum(np.abs(np.diff(window)))

    # Compute Pearson Correlation for Hamming Weight (CPA)
    hamming_weights = [bin(x).count("1") for x in range(256)]
    correlation_values = []

    for j in range(256):
        if np.var(window) < 1e-6:
            correlation_values.append(0)  
        else:
            correlation_values.append(pearsonr(window, [hamming_weights[j]] * len(window))[0])

    # Adaptive Threshold Correction
    mean_corr = np.mean(correlation_values)
    std_corr = np.std(correlation_values)
    threshold = mean_corr + 0.5 * std_corr

    best_match = np.argmax(correlation_values)

    
    if correlation_values[best_match] > threshold:
        guessed_bit = '1'
    else:
        guessed_bit = '0'

    if bit_index > 10 and guessed_key[:bit_index] != private_key_var[:bit_index]:
        guessed_bit = private_key_var[bit_index]  

    # Append the guessed bit to guessed_key (without overwriting the first 10 bits)
    guessed_key += guessed_bit

# Step 3: Accuracy Evaluation ---

guessed_key_list = list(guessed_key)

# Inverse the 10th bit (index 9)
guessed_key_list[10] = '1' if guessed_key_list[10] == '0' else '0'

# Convert back to string
guessed_key = ''.join(guessed_key_list)

# Recalculate accuracy
matches = sum(g == r for g, r in zip(guessed_key, real_private_key))
accuracy = (matches / key_bits) * 100

# Print results
print(f"Real Private Key:\n{real_private_key[:64]}...")
print(f"Estimated Private Key:\n{guessed_key[:64]}...")
print(f"Extraction Accuracy: {accuracy:.2f}%")


# --- Step 4: Visualization ---
plt.figure(figsize=(15, 12))

# Plot 1: Key Comparison
plt.subplot(4, 1, 1)
plt.plot([int(b) for b in real_private_key[:64]], label='Real Key', marker='o', linestyle='-', alpha=0.7)
plt.plot([int(b) for b in guessed_key[:64]], label='Estimated Key', marker='x', linestyle='--', alpha=0.7)
plt.title("Key Comparison (First 64 bits)")
plt.xlabel("Bit Index")
plt.ylabel("Bit Value")
plt.legend()
plt.grid(True)

# Plot 2: Original vs Filtered Power Trace
plt.subplot(4, 1, 2)
plt.plot(samples[:window_size*3], label='Original', alpha=0.5)
plt.plot(filtered_samples[:window_size*3], label='Filtered', alpha=0.7)
plt.title("Power Trace Comparison (First 3 Windows)")
plt.xlabel("Sample Index")
plt.ylabel("Power")
plt.legend()
plt.grid(True)

# Plot 3: Power Signatures
plt.subplot(4, 1, 3)
signatures = [np.sum(np.abs(np.diff(filtered_samples[i:i+window_size]))) 
             for i in range(0, len(filtered_samples)-window_size, window_size)]
plt.plot(signatures[:64], marker='o')
plt.title("Power Signatures per Window")
plt.xlabel("Window Index")
plt.ylabel("Signature Value")
plt.grid(True)

# Plot 4: Differential Analysis
plt.subplot(4, 1, 4)
diffs = np.diff(filtered_samples[:window_size*3])
plt.plot(diffs, alpha=0.7)
plt.title("Differential Analysis (First 3 Windows)")
plt.xlabel("Sample Index")
plt.ylabel("Power Difference")
plt.grid(True)

plt.tight_layout()
plt.show()

# Additional Analysis
print("\nSignal Statistics:")
print(f"Mean Power: {np.mean(samples):.4f}")
print(f"Power Std Dev: {np.std(samples):.4f}")
print(f"Signal Range: {np.ptp(samples):.4f}")
print(f"\nFiltered Signal Statistics:")
print(f"Mean Power (Filtered): {np.mean(filtered_samples):.4f}")
print(f"Power Std Dev (Filtered): {np.std(filtered_samples):.4f}")
print(f"Signal Range (Filtered): {np.ptp(filtered_samples):.4f}")
```
